## <font color='264CC7'> Introducción </font>

Este notebook está diseñado para poner en práctica los conceptos de bases de datos relacionales sobre datos reales.

Los paquetes necesarios son:

```python
# Paquetes necesarios
import pandas as pd
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
import numpy as np
```

---
## <font color='264CC7'> Datos </font>


Para los ejercicios se usará el conjunto de datos «120 years of Olympic history: athletes and results», disponible en [Kaggle](https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results). Este conjunto de datos contiene información sobre los atletas, equipos y eventos de los Juegos Olímpicos desde 1896 hasta 2016.

```python
# Crear las tablas
tabla_atletas = pd.read_csv("https://raw.githubusercontent.com/andres-merino/PracticaBasesRalacionales/refs/heads/main/data/athlete_events.csv")
tabla_paises = pd.read_csv("https://raw.githubusercontent.com/andres-merino/PracticaBasesRalacionales/refs/heads/main/data/noc_regions.csv")

# Mostrar las tablas
print("Tabla de atletas")
display(tabla_atletas.head())
print("Tabla de paises")
display(tabla_paises.head())
```

### <font color='264CC7'> Consultas </font>


#### <font color='264CC7'> Consulta 1 </font>

Consulta todos los juegos olímpicos que se han celebrado.

```python
# Consulta: todos los juegos olímpicos que se han celebrado.
tabla_atletas["Games"].unique()
```

```python
# Consulta SQL: todos los juegos olímpicos que se han celebrado.
consulta = """
SELECT DISTINCT Games
FROM tabla_atletas
"""
pysqldf(consulta)
```

#### <font color='264CC7'> Consulta 2 </font>

Consulta el número de juegos olímpicos que se han celebrado.

```python
# Consulta: el número de juegos olímpicos que se han celebrado.
tabla_atletas["Games"].nunique()
```

```python
# Consulta SQL: el número de juegos olímpicos que se han celebrado.
consulta = """
SELECT COUNT(DISTINCT Games) AS Num_Juegos_Olimpicos
FROM tabla_atletas
"""
pysqldf(consulta)
```

#### <font color='264CC7'> Consulta 3 </font>

Consulta el número total de países que participaron en cada juego olímpico.

```python
# Consulta: úmero total de países que participaron en cada juego olímpico
display(tabla_atletas.groupby("Games")["NOC"].nunique())
```

```python
# Consulta SQL: Jugadores de los equipos
consulta = """
SELECT Games, COUNT(DISTINCT NOC) AS Total_Paises
FROM tabla_atletas
GROUP BY Games;
"""
pysqldf(consulta)
```
